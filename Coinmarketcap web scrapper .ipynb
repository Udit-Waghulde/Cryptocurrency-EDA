```python
import pandas as pd 
import json
from bs4 import BeautifulSoup 
import csv 
import requests 
```

```python
def req_to_csv(r, name):
    res = []
    Data = r.json()
    for d in Data['data']['quotes']:
        res.append(d)
    df = pd.json_normalize(res)
    df.to_csv(name)
```
